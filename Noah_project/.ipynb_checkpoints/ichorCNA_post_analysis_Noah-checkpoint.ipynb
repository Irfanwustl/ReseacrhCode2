```python
import os
import pandas as pd
#results_dir = "/logo2/irfan/BioInfoSoftware/ichorCNA/scripts/snakemake/results/ichorCNA"

results_dir = "/logo2/irfan/BioInfoSoftware/ichorCNA/scripts/snakemake/results_notusedLowTumorSettings/ichorCNA"
```

```python


def extract_ichorcna_data(results_dir):
    """
    Extracts relevant data from ichorCNA results' params.txt files.

    Parameters:
    - results_dir (str): Path to the directory containing ichorCNA results folders.

    Returns:
    - DataFrame: A Pandas DataFrame containing extracted data.
    """

    # Columns for the resulting DataFrame
    columns = ["sample_name", "tumor_fraction", "ploidy", "gender", "ChrY_coverage_fraction", "ChrX_median_log_ratio"]
    
    # Initialize a list to store extracted data dictionaries
    data_list = []
    
    # Iterating through each folder in the results directory
    for sample_folder in os.listdir(results_dir):
        sample_path = os.path.join(results_dir, sample_folder)
        
        # Confirming that it's a folder
        if os.path.isdir(sample_path):
            # Searching for the params.txt file with .params.txt extension
            for file_name in os.listdir(sample_path):
                if file_name.endswith(".params.txt"):
                    params_file_path = os.path.join(sample_path, file_name)
                    break  # Exit the loop once we find the first .params.txt file
            else:  # No params.txt file found
                print(f"No params.txt file found for sample: {sample_folder}")
                continue
            
            # Extracting data from params.txt
            with open(params_file_path, "r") as file:
                # Initialize a dictionary to store extracted data
                data = {"sample_name": sample_folder}
                
                # Iterating through each line in params.txt to extract data
                for line in file:
                    line = line.strip()  # Remove leading/trailing whitespace
                    if "Gender:" in line:
                        data["gender"] = line.split(":")[1].strip()
                    elif "Tumor Fraction:" in line:
                        data["tumor_fraction"] = float(line.split(":")[1].strip())
                    elif "Ploidy:" in line:
                        data["ploidy"] = float(line.split(":")[1].strip())
                    elif "ChrY coverage fraction:" in line:
                        data["ChrY_coverage_fraction"] = float(line.split(":")[1].strip())
                    elif "ChrX median log ratio:" in line:
                        data["ChrX_median_log_ratio"] = float(line.split(":")[1].strip())
                
                # Adding extracted data to the data list
                data_list.append(data)
                    
    # Convert the list of data dictionaries to a DataFrame
    result_df = pd.DataFrame(data_list, columns=columns)
    
    return result_df
```

```python
ichorcna_data = extract_ichorcna_data(results_dir)
```

```python
ichorcna_data['sample_name'] = ichorcna_data['sample_name'].str.replace('-auto.final.dedup.bam', '',regex=False)
ichorcna_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_rel, wilcoxon

def paired_plot(dataframe):
    # Filter and separate data
    filtered_data = dataframe[dataframe['sample_name'].str.startswith('H')]
    before_sonobiopsy = filtered_data[filtered_data['sample_name'].str.contains('_B01')].copy()
    after_sonobiopsy = filtered_data[filtered_data['sample_name'].str.contains('_B03')].copy()

    # Extract the base sample name (e.g., "HO1") from the "sample_name" column for pairing
    before_sonobiopsy['base_sample'] = before_sonobiopsy['sample_name'].str.extract(r'(\w{3})')
    after_sonobiopsy['base_sample'] = after_sonobiopsy['sample_name'].str.extract(r'(\w{3})')
    
    # Merge the two DataFrames based on this base sample name for correct pairing
    paired_df = pd.merge(before_sonobiopsy, after_sonobiopsy, on='base_sample', suffixes=('_before', '_after'))
    display(paired_df)

    # Create a long-form DataFrame for plotting
    plot_df = pd.melt(paired_df, id_vars='base_sample', value_vars=['tumor_fraction_before', 'tumor_fraction_after'], 
                      var_name='time', value_name='tumor_fraction')
    plot_df['time'] = plot_df['time'].map({'tumor_fraction_before': 'Before Sonobiopsy', 'tumor_fraction_after': 'After Sonobiopsy'})

    # Conduct paired t-test
    t_stat, t_p_value = ttest_rel(paired_df['tumor_fraction_before'], paired_df['tumor_fraction_after'])
    
    # Conduct Wilcoxon signed-rank test
    w_stat, w_p_value = wilcoxon(paired_df['tumor_fraction_before'], paired_df['tumor_fraction_after'])

    # Plotting
    plt.figure(figsize=(10, 6))
    
    # Using pointplot for the mean and line to show connection between pairs
    sns.pointplot(x='time', y='tumor_fraction', data=plot_df, join=True, errorbar=None, markers='d', scale=1.5, color='black')

    for i in range(len(paired_df)):
        color = 'red' if paired_df['tumor_fraction_before'].iloc[i] > paired_df['tumor_fraction_after'].iloc[i] else 'green'
        plt.plot([0, 1], [paired_df['tumor_fraction_before'].iloc[i], paired_df['tumor_fraction_after'].iloc[i]], color=color, alpha=0.5)

    plt.title(f"Tumor Fraction Comparison (T-test p-value: {t_p_value:.4f}, Wilcoxon p-value: {w_p_value:.4f})")
    plt.ylabel('Tumor Fraction')
    plt.show()
    
    return t_stat, t_p_value, w_stat, w_p_value






# You would call this function and pass your DataFrame, like this:
# plot_with_sample_names(combined_data)


# Using the function on ichorcna_data
t_stat, t_p_value, w_stat, w_p_value = paired_plot(ichorcna_data)
print(f"T-test: T-statistic = {t_stat}, P-value = {t_p_value}")
print(f"Wilcoxon: Statistic = {w_stat}, P-value = {w_p_value}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import mannwhitneyu

def box_plot_with_samples(dataframe, group1, group2):
    # Function to filter data based on group name
    def filter_group(group):
        if group == "before_sonobiopsy":
            return dataframe[dataframe['sample_name'].str.contains('_B01')].copy()
        elif group == "after_sonobiopsy":
            return dataframe[dataframe['sample_name'].str.contains('_B03')].copy()
        elif group == "healthy":
            return dataframe[dataframe['sample_name'].str.startswith('NU') | dataframe['sample_name'].str[0].str.isdigit()].copy()
        else:
            return pd.DataFrame()

    
    # Filter data based on the two group names provided
    group1_data = filter_group(group1)
    group2_data = filter_group(group2)
    
    # Combining the two groups
    group1_data['group'] = group1
    group2_data['group'] = group2
    combined_data = pd.concat([group1_data, group2_data])
    
    # Mann-Whitney U test
    mwu_stat, mwu_p_value = mannwhitneyu(group1_data['tumor_fraction'], group2_data['tumor_fraction'])
    
    # Plotting
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='group', y='tumor_fraction', data=combined_data, palette="coolwarm", showfliers=False) # showfliers=False to hide outliers since we are overlaying them using swarmplot
    sns.swarmplot(x='group', y='tumor_fraction', data=combined_data, color='black', size=5, alpha=0.7)
    plt.title(f"Tumor Fraction Comparison (MWU test p-value: {mwu_p_value:.4f})")
    plt.ylabel('Tumor Fraction')
    plt.show()
    
    return mwu_stat, mwu_p_value

# Usage example:
mwu_stat, mwu_p_value = box_plot_with_samples(ichorcna_data, "before_sonobiopsy", "after_sonobiopsy")
print(f"Mann-Whitney U test: Statistic = {mwu_stat}, P-value = {mwu_p_value}")

mwu_stat, mwu_p_value = box_plot_with_samples(ichorcna_data, "before_sonobiopsy", "healthy")
print(f"Mann-Whitney U test: Statistic = {mwu_stat}, P-value = {mwu_p_value}")

mwu_stat, mwu_p_value = box_plot_with_samples(ichorcna_data, "after_sonobiopsy", "healthy")
print(f"Mann-Whitney U test: Statistic = {mwu_stat}, P-value = {mwu_p_value}")
```
