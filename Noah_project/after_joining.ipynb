```python
import pandas as pd


bedfile = 'all_bedswithHealthy_bg_all_matrix_DMRs.bed'


with open(bedfile, 'r') as file:
    first_line = file.readline().strip()
    # Split the first line into column names assuming tab-delimited
    column_names = first_line.split('\t')

    
additional_columns = ['DMRchr', 'DMRstart', 'DMRstop']
column_names = column_names + additional_columns


intersection_df = pd.read_csv(bedfile, sep='\t', header=None, skiprows=1)

intersection_df.columns = column_names
display(intersection_df.head())
```

```python
# Remove a fixed prefix 'prefix_' from the column names
intersection_df.columns = intersection_df.columns.str.replace('^Sample_', '', regex=True)

# Remove a fixed suffix '_suffix' from the column names
intersection_df.columns = intersection_df.columns.str.replace('_CTF.bed$', '', regex=True)

# Display the updated DataFrame columns
intersection_df.head()
```

```python

# Load the Excel file
clinical_df = pd.read_excel('short CTF Clinical Sheet_Irfan_3-2-24_irf.xlsx')

# Segment the DataFrame based on 'Histology simplified' and extract column names for each condition
adenocarcinoma_cols = clinical_df[clinical_df['Histology symplified'] == 'Adenocarcinoma'].iloc[:, 0].tolist()
scc_cols = clinical_df[clinical_df['Histology symplified'] == 'SCC'].iloc[:, 0].tolist()
benign_cols = clinical_df[clinical_df['Histology symplified'] == 'Benign'].iloc[:, 0].tolist()
healthy_cols = clinical_df[clinical_df['Histology symplified'] == 'Healthy'].iloc[:, 0].tolist()



# Combine the lists in the desired order
columns_to_keep_ordered = adenocarcinoma_cols + scc_cols + benign_cols + healthy_cols

#print(columns_to_keep_ordered)
# Add the additional columns at the beginning of the list
additional_columns = ['chrom', 'start', 'end', 'DMRchr', 'DMRstart', 'DMRstop']
columns_to_keep = additional_columns + columns_to_keep_ordered

# Filter the intersection_df to include only the ordered columns
# Ensure that the columns in columns_to_keep exist in intersection_df to avoid KeyError
valid_columns = [col for col in columns_to_keep if col in intersection_df.columns]
#print(valid_columns)

intersection_df_filtered = intersection_df[valid_columns]


print(intersection_df_filtered.shape)
# Display the resulting DataFrame
display(intersection_df_filtered.head())
```

# PCA

```python
# Drop the 'chrom', 'start', 'end' columns from the DataFrame
intersection_df_filtered = intersection_df_filtered.drop(['chrom', 'start', 'end'], axis=1)

# Display the resulting DataFrame to confirm the columns are dropped
display(intersection_df_filtered.head())
```

```python
# Group by 'DMRchr', 'DMRstart', 'DMRstop' and calculate the mean for each group
average_df = intersection_df_filtered.groupby(['DMRchr', 'DMRstart', 'DMRstop'], as_index=False).mean()


print(average_df.shape)
# Display the resulting DataFrame
display(average_df.head())
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'average_df' is your DataFrame

# Step 1: Set 'DMRchr', 'DMRstart', 'DMRstop' as the index
average_df_indexed = average_df.set_index(['DMRchr', 'DMRstart', 'DMRstop'])

# Step 2: Transpose the DataFrame
transposed_df = average_df_indexed.transpose()

# Step 3: Handle missing values. Options include dropping rows/columns with NAs or imputing values.
# Here, we'll impute missing values with the mean of each column (feature)
transposed_df_imputed = transposed_df.fillna(transposed_df.mean())

# Step 4: Normalize the data
scaler = StandardScaler()
transposed_df_normalized = scaler.fit_transform(transposed_df_imputed)

# Step 5: Perform PCA
pca = PCA(n_components=2)  # Example: reduce to 2 dimensions for visualization purposes
pca_result = pca.fit_transform(transposed_df_normalized)

# Convert PCA result to DataFrame for easier handling
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# If you want to include the index (patient IDs) back into the PCA result DataFrame:
pca_df.index = transposed_df_imputed.index

plt.figure(figsize=(10, 7))
plt.scatter(pca_df['PC1'], pca_df['PC2'])

plt.title('PCA of Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Optionally, add annotations or labels for each point
for i, txt in enumerate(pca_df.index):
    plt.annotate(txt, (pca_df['PC1'][i], pca_df['PC2'][i]))

plt.show()
```

```python
histology_labels = clinical_df.set_index('PatientID')['Histology symplified']  # Adjust 'PatientID' as necessary

# Add the histology labels to the pca_df
pca_df['Histology'] = histology_labels.loc[pca_df.index].values

# Plot the PCA results with histology labels
plt.figure(figsize=(10, 7))

# Plot each histology category in a different color
for label, color in [('Adenocarcinoma', 'brown'), ('SCC', 'red'), ('Benign','blue' ),('Healthy', 'green')]:
    condition = pca_df['Histology'] == label
    plt.scatter(pca_df[condition]['PC1'], pca_df[condition]['PC2'], label=label, color=color)

plt.title('PCA of Dataset by Histology')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

plt.show()
```

```python


def reorder_rows_based_on_df_index(df_to_order: pd.DataFrame, df_order_reference: pd.DataFrame) -> pd.DataFrame:
    """
    Reorder the rows of df_to_order based on the order of df_order_reference's index.
    
    Parameters:
    - df_to_order: DataFrame whose rows need to be reordered.
    - df_order_reference: DataFrame or Series whose index indicates the desired order of rows.
    
    Returns:
    - DataFrame with rows reordered as per df_order_reference's index.
    """
    # Use df_order_reference's index to reorder rows in df_to_order
    reordered_df = df_to_order.loc[df_order_reference.index]
    
    return reordered_df
```

```python
average_df_indexed_hypo_ordered = reorder_rows_based_on_df_index(average_df_indexed, hypo_sorted)
average_df_indexed_hyper_ordered = reorder_rows_based_on_df_index(average_df_indexed, hyper_sorted)

display(average_df_indexed_hypo_ordered.head())
display(average_df_indexed_hyper_ordered.head())
```
